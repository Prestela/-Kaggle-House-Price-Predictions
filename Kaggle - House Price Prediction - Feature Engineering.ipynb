```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler

pd.set_option('display.max_columns', None)
%matplotlib inline
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
test_y = pd.read_csv('sample_submission.csv')
df = train
train.shape
test = test.merge(test_y, left_on='Id', right_on='Id')
test
```

**Categorise variables into various data types**

```python
num_vars = [var for var in df.columns if (df[var].dtype!='O')&(var not in ['Id'])]
num_vars
```

```python
cat_vars = [var for var in df.columns if df[var].dtype=='O']
```

```python
def elapsed_years(df, var):
    return df[var]-df['YrSold']
```

```python
for var in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    df[var] = elapsed_years(df,var)
```

```python
temporal_vars = ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']
```

```python
discrete_vars = [var for var in num_vars if df[var].nunique()<=20 not in temporal_vars+['Id']]
```

```python
continuous_vars = [var for var in num_vars if df[var].nunique()>20 not in temporal_vars+['Id']]
```

```python
df[continuous_vars].isnull().sum()
```

**Fill in missing variables**

```python
# fill missing values
for var in discrete_vars:
    print(f'{var} : {df[var].median()}')
```

```python
train[discrete_vars].dtypes
```

```python
vars_with_na=[var for var in train.columns if (train[var].isnull().sum()>0) & (train[var].dtype!='O')]
```

```python
for var in num_vars:
    median_val = train[var].median()
    train[var+'_na'] = np.where(train[var].isnull(),1 ,0)
    test[var+'_na'] = np.where(test[var].isnull(),1 ,0)
    
    train[var] = train[var].fillna(median_val)
    test[var] = test[var].fillna(median_val)
```

```python
train[discrete_vars].isnull().sum()
test[discrete_vars].isnull().sum()
```

```python
train[continuous_vars].isnull().sum()
#test[continuous_vars].isnull().sum()
```

```python
for var in continuous_vars:
    train[var] = train[var].fillna(train[var].mean())
    continuous_vars_test = [var for var in train[continuous_vars] if var != 'SalePrice']
for var in continuous_vars_test:
    test[var] = test[var].fillna(test[var].mean())
```

```python
for var in temporal_vars:
    train[var] = train[var].fillna(train[var].mean())
    continuous_vars_test = [var for var in train[continuous_vars] if var != 'SalePrice']
for var in continuous_vars_test:
    test[var] = test[var].fillna(test[var].mean())
```

```python
train[cat_vars] = train[cat_vars].fillna('Missing')
test[cat_vars] = test[cat_vars].fillna('Missing')
```

```python
train[cat_vars].isnull().sum()
```

**Replace cat variables labels with ordinal labels**

```python
def analyse_cat_vars(df, var, rare_perc = 0.05):
    tmp = df.groupby(var)['SalePrice'].count()/len(df)
    print(tmp[tmp>0.05].index)
    print('\n')
```

```python
for var in cat_vars:
    analyse_cat_vars(df, var)
```

```python
def frequent_label(df, var, rare_perc=0.05):
    tmp = df.groupby(var)['SalePrice'].count()/len(df)
    return tmp[tmp>rare_perc].index
```

```python
for var in cat_vars:
    frequent_ls = frequent_label(train, var)
    train[var]=np.where(train[var].isin(frequent_ls), train[var], "Rare")
    test[var]=np.where(test[var].isin(frequent_ls), test[var], "Rare")
```

```python
for var in cat_vars:
    analyse_cat_vars(train,var)
```

```python
train['MSZoning'].unique()
```

```python
def ordinal_labels(df, var, rare_perc=0.05):
    label= df.groupby(var)['SalePrice'].mean().sort_values().index
    
    new_label ={k: i for i,k in enumerate(label)}
    return new_label
```

```python
train[cat_vars].isnull().sum()
```

```python
for var in cat_vars:
    label = ordinal_labels(train, var)
    #print(var)
    #print(label)
    train[var] = train[var].map(label)
    test[var] = test[var].map(label)
```

```python
train[cat_vars].head()
```

```python
train_vars = [var for var in train.columns if var not in ['Id', 'SalePrice']]
test_vars = [var for var in test.columns if var not in ['Id']]
```

```python
train[train_vars].isnull().sum()
```

**Transform variables**

```python
scaler = MinMaxScaler()
scaler.fit(train[train_vars])
train[train_vars] = scaler.transform(train[train_vars])
```

```python
test['Id']=test['Id'].astype('float')
test['Id'].dtype
```

```python
scaler.fit(test[test_vars])
```

```python
test[test_vars] = scaler.transform(test[test_vars])
```
