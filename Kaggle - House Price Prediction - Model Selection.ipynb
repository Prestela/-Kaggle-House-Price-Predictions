```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import Lasso, Ridge, LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

from xgboost import XGBRegressor

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam

pd.pandas.set_option('display.max_columns', None)
```

```python
selected_features = pd.read_csv('Selected_features.csv',header=None)
features = selected_features[0].tolist()
features
```

```python
train = pd.read_csv('New_Train.csv')
```

```python
test = pd.read_csv('New_Test.csv')
test
```

```python
linear_model = LinearRegression()
```

```python
X_train=train[features].to_numpy()
X_test=test[features].to_numpy()
```

```python
y_train=train["SalePrice"].to_numpy()
y_test=test["SalePrice"].to_numpy()
```

***Linear Regression Model***

```python
linear_model.fit(X_train,y_train)
```

```python
y_pred_lm = linear_model.predict(X_test)
y_pred_lm.shape
```

```python
linear_model.coef_
```

```python
mse_1 = mean_squared_error(y_test, y_pred_lm)
```

```python
r2_1 = r2_score(y_test, y_pred_lm)
```

***XGBoost***

```python
xgboost = XGBRegressor()
```

```python
xgboost.fit(X_train,y_train)
```

```python
y_pred_xgb = xgboost.predict(X_test)
y_pred_xgb
```

```python
mse_1
```

***Neural Network***

```python
model = Sequential()
model.add(Dense(6, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='Adam', loss='mse')
```

```python
model.fit(x=X_train, y=y_train,
         validation_data=(X_test, y_test),
         batch_size=200, epochs=400)
```

```python
y_pred_nn=model.predict(X_test)
y_pred_nn
```

```python
mse_3 = model.evaluate(X_test,y_test)
```

```python
mse_3
np.var(y_pred_nn)
```

```python
print(f'Linear Regression: {mse_1}')
print(f'XGBoost: {mse_2}')
print(f'Neural Network: {mse_3}')
```
